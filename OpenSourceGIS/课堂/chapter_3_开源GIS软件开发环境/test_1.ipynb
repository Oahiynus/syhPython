```python
def num_higher(s,s_high):
    num=0
    for a in s:
        if a >= s_high:
            num = num+1
    return num
```

```python
scores = [93,60,76,56,98,65,89,78,90,73,56,98,78,90,73,78,90,73]
num=0
for a in scores:
    if a >= 90:
        num = num+1
print(num)
```

```python
num_high = num_higher(s=scores, s_high=90)
print(num_high)
```
